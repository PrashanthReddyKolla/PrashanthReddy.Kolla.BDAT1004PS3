## Question 1: Occuaptions

### Step 1. Import the necessary libraries

```python
import pandas as pd
import seaborn
import numpy as np
```

### Step 2. Import the dataset from this address.

```python
url1 = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
```

### Step 3. Assign it to a variable called users

```python
users = pd.read_csv(url1, sep = "|")
```

### Step 4. Discover what is the mean age per occupation

```python
age_per_occupation = users.groupby('occupation').mean()
round(age_per_occupation['age'], 2)
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
each_gen_count = users.groupby(['gender', 'occupation'])['gender'].count()  #Count of each occupation grouped by gender
men_count      = each_gen_count['M':]                                       #Count of men for each occupation
total_count    = users.groupby(['occupation'])['gender'].count()            #Total count for each occupation
men_ratio_occ  = (men_count/total_count).round(2)                           #Men ratio for each occupation
print(men_ratio_occ.sort_values(ascending=False))                           #Sorting the ratios in descending order
```

### Step 6. For each occupation, calculate the minimum and maximum ages
### Minimum Age for each Occupation

```python
min_age_occupation = users.groupby(['occupation']).min()
min_age_occupation['age']
```

### Maximum age for each occupation

```python
max_age_occupation = users.groupby(['occupation']).max()
max_age_occupation['age']
```

### Step 7. For each combination of occupation and sex, calculate the mean age

```python
occ_gen_mean = users.groupby(['occupation', 'gender']).mean()
round(occ_gen_mean['age'], 2)
```

### Step 8. For each occupation present the percentage of women and men

```python
gen_count   = users.groupby(['occupation', 'gender'])['gender'].count()         #Count of gender grouped by occupation
total_count = users.groupby(['occupation'])['gender'].count()                   #Total count
Percentage  = ((gen_count/total_count)*100).round(2)                            #% for each gender grouped by occupation
print(Percentage)
```

### Step 1. Import the necessary libraries

```python
import pandas as pd
import seaborn
import datetime
import numpy as np
```

### Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index.

```python
data    = pd.read_csv(url4, sep = "\s+")
NewDate = data['Yr'].astype(str) + '-'+ data['Mo'].astype(str) + '-' + data['Dy'].astype(str)
data.insert(0, 'Date', NewDate)
data['Date'] =  pd.to_datetime(data['Date'])
data = data.drop(columns = ['Yr', 'Mo', 'Dy'])
data
```

### Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

```python
#Function
def valid_year_check(date_column):
    
    if date_column.year > 2000:
        year = date_column.year - 100
    else:
        year = date_column.year
    
    return datetime.date(year,date_column.month,date_column.day)

#Replacing invalid years with valid values
data['Date'] = data['Date'].apply(valid_year_check)
data
```

### Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].

### Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. Label both lines and show the legend.

```python
import matplotlib.pyplot as plt
import pandas as pd

us_marr_div = pd.read_csv("C:/Users/user/Desktop/BDAT-2020/1004 - Data Programming/ProblemSet3/us-marriages-divorces-1867-2014.csv")

plt.plot(us_marr_div.Year, us_marr_div['Marriages'], label='Marriages')
plt.plot(us_marr_div.Year, us_marr_div['Divorces'], label='Divorces')

plt.title("Marriages and Divorces in the U.S between 1867-2014")
plt.xlabel("Year")
plt.ylabel(" Number of Marriages/Divorces");

plt.legend();
```

### Create a pie chart showing the fraction of all Roman Emperors that were assassinated.
### Make sure that the pie chart is an even circle, labels the categories, and shows the percentage breakdown of the categories.

```python
import pandas as pd
import matplotlib.pyplot as plt

roman_emperors = pd.read_csv("C:/Users/user/Desktop/BDAT-2020/1004 - Data Programming/ProblemSet3/roman-emperor-reigns.csv")

death_group = roman_emperors.groupby('Cause_of_Death')['Emperor'].count()

death_group.plot.pie(y='Cause_of_Death', figsize=(10, 10), autopct='%1.1f%%', startangle=0, textprops={'fontsize': 10})

plt.axis('equal')
plt.title('Causes of deaths of Roman Emperors')
plt.ylabel('')
```
